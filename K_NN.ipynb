FEATURE SCALING : Transformasi dan Normalisasi Data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

MODELLING DENGAN KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

#definisikan nilai K dan buat KNN Regressor
k = 3
knn = KNeighborsRegressor(n_neighbors=k)

#fitkan midek KNN ke data train
knn.fit(X_train, y_train)
```

```python
#prediksi pada datatrain dan data test
y_pred_train = knn.predict(X_train)
y_pred_test = knn.predict(X_test)
```

```python
# Menghitung MSE untuk data train dan data test
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
```

```python
print("MSE data train:", mse_train)
print("MSE data train:", mse_train)
```

Hasil prediksi berdasarkan perhitungan mean squared error

```python
import matplotlib.pyplot as plt

#plot data train
plt.scatter(y_train, y_pred_train)
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'k--', lw=3)
plt.xlabel('Nilai aktual')
plt.ylabel('Nilai akrediksi')
plt.title('Data train: MSE = %.2f' %mse_train)
plt.show()

#plot data test
plt.scatter(y_test, y_pred_test)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=3)
plt.xlabel('Nilai aktual')
plt.ylabel('Nilai akrediksi')
plt.title('Data train: MSE = %.2f' %mse_test)
plt.show()


```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

Selanjutnya kodingan untuk menghitung jarak antara data poin dan menemukan tetangga terdekat:

```python
# Memilih kolom yang akan digunakan sebagai data poin
X = df_train[['infoavail', 'happy']].values
```

```python
# Membuat dan melatih model K-NN
n_neighbors = 2
knn_model = NearestNeighbors(n_neighbors=n_neighbors)
knn_model.fit(X)
```

```python
# Data poin baru yang akan dicari tetangganya
new_housecost = 800000
new_events = 5
new_point = np.array([[new_housecost, new_events]])
```

```python
# Menghitung jarak dan indeks tetangga terdekat
distances, indices = knn_model.kneighbors(new_point)
```

```python
# Menampilkan hasil
print("Data poin baru:", new_point)
print(f"{n_neighbors} tetangga terdekat:")
for i in range(n_neighbors):
    print(f"Index: {indices[0, i]}, Jarak: {distances[0, i]}")
```
